```python
import os
import pandas as pd
import numpy as np
from datasets import Dataset
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain_text_splitters import RecursiveCharacterTextSplitter
from ragas import evaluate
try:
    from ragas.metrics import (
        faithfulness,
        answer_relevancy,
        context_precision,
        context_recall,
    )
except ImportError:
    from ragas.metrics.collections import (
        faithfulness,
        answer_relevancy,
        context_precision,
        context_recall,
    )

# APIキー設定
import dotenv
dotenv.load_dotenv()

key = os.getenv("OPENAI_API_KEY_LLUMINAI")
os.environ["OPENAI_API_KEY"] = key

# ==========================================
# 1. 実験設定：文書と正解のペア
# ==========================================
experiments = [
    {
        "type": "1. Short",
        "text": text_short,
        "q": "少女と青年が持っていた日記には、どんな共通点がありましたか？",
        "truth": "明日の日付で「今日、運命の人に会う」と同じ文字で続きが書かれていた。",
    },
    {
        "type": "2. Medium",
        "text": text_medium,
        "q": "祖父が「唯一の心残り」だとしていたことは何ですか？",
        "truth": "ユキとの約束を果たせなかったこと。",
    },
    {
        "type": "3. Long",
        "text": text_long,
        "q": "祖母が灯台の木箱の中に隠していた、「もう一つの小さな包み」の中身は何でしたか？",
        "truth": "隆が描いた、若い日の祖母の肖像画。",
    },
    {
        "type": "4. Very Long",
        "text": text_very_long,
        "q": "郵便配達員の「私」は、白石幸子さんに手紙を届けるために、最終的にどこへ向かいましたか？", 
        "truth": "北海道の函館にある病院（または函館市内）。",
    },
]

# 実験するチャンクサイズ（6種類）
chunk_sizes = [128, 256, 512, 1024, 2048, 4096]

# ==========================================
# ヘルパー関数: スコアを安全に取り出す
# ==========================================
def get_safe_score(scores, key):
    """
    Ragasの結果オブジェクトから数値を安全に取り出す関数
    """
    try:
        val = scores[key]
    except (KeyError, TypeError):
        return 0.0
    
    # リストの場合（[1.0]など）は中身を取り出す
    if isinstance(val, list):
        if len(val) > 0:
            val = val[0]
        else:
            return 0.0
            
    # NaN（計算不能）の場合
    if pd.isna(val):
        return 0.0
        
    return float(val)

# ==========================================
# 2. 総当たり評価ループ
# ==========================================
results_log = []
print("🧪 Ragas深掘り実験を開始します（4指標計測）...")

llm = ChatOpenAI(model="gpt-4o-mini")
embeddings = OpenAIEmbeddings()

# 評価指標リスト
metrics = [faithfulness, answer_relevancy, context_precision, context_recall]

for exp in experiments:
    print(f"\n📂 Document Type: {exp['type']}")
    
    for chunk_size in chunk_sizes:
        # --- A. インデックス構築 ---
        splitter = RecursiveCharacterTextSplitter(
            chunk_size=chunk_size,
            chunk_overlap=int(chunk_size * 0.1),
            length_function=len,
        )
        docs = splitter.create_documents([exp['text']])
        if not docs: docs = splitter.create_documents([exp['text']])
        vectorstore = FAISS.from_documents(docs, embeddings)
        
        # --- B. RAG検索 & 回答生成 ---
        retriever = vectorstore.as_retriever(search_kwargs={"k": 3})
        retrieved_docs = retriever.invoke(exp["q"])
        contexts = [d.page_content for d in retrieved_docs]
        
        context_str = "\n".join(contexts)
        prompt = f"""以下の[文脈]のみに基づいて、[質問]に答えてください。
文脈にない情報は使わないでください。

[文脈]:
{context_str}

[質問]:
{exp['q']}
"""
        ans = llm.invoke(prompt).content
        
        # --- C. Ragas評価 ---
        dataset = Dataset.from_dict({
            "question": [exp["q"]],
            "answer": [ans],
            "contexts": [contexts],
            "ground_truth": [exp["truth"]]
        })
        
        scores = evaluate(
            dataset=dataset,
            metrics=metrics,
            llm=llm, 
            embeddings=embeddings,
            raise_exceptions=False
        )
        
        # 安全に値を取得
        s_faith = get_safe_score(scores, 'faithfulness')
        s_ans_rel = get_safe_score(scores, 'answer_relevancy')
        s_cont_prec = get_safe_score(scores, 'context_precision')
        s_cont_rec = get_safe_score(scores, 'context_recall')
        
        print(f"   - Chunk {chunk_size:4}: Recall={s_cont_rec:.2f}, Faith={s_faith:.2f}")
        
        results_log.append({
            "Type": exp["type"],
            "Chunk": chunk_size,
            "Context Recall": s_cont_rec,
            "Context Precision": s_cont_prec,
            "Faithfulness": s_faith,
            "Answer Relevancy": s_ans_rel
        })

# ==========================================
# 3. 結果の集計と表示
# ==========================================
df = pd.DataFrame(results_log)

print("\n" + "="*50)
print("🏆 実験結果：指標別ヒートマップ")
print("="*50)

metric_names = ["Context Recall", "Context Precision", "Faithfulness", "Answer Relevancy"]

for m in metric_names:
    print(f"\n📊 {m} (高いほど良い)")
    print("-" * 30)
    try:
        pivot = df.pivot(index="Type", columns="Chunk", values=m)
        print(pivot.round(2))
    except Exception as e:
        print(f"集計エラー: {e}")

print("\n✅ 実験完了")
```
